```python
# Function to extract manufacturer from description
def extract_manufacturer(model):
    if pd.isna(model):
        return None
    # Look for common manufacturer names in the description
    manufacturers = ["Mazda", "Ford", "Toyota", "Honda", "Volvo", "Kia", "Acura", "Ram", "Jeep","Plymouth",
                     "Maserati","Scion","Oldsmobile","Smart","lac","Caterpillar","Mercedes","Freightliner",
                     "Nova","Sterling","Suzuki","Mercedes-Benz","Volkswagen","Bentley","Hummer","International 4700",
                     "Peterbilt","Poniac","Saab","Isuzu","International","ABC","Mack","Wheego","Coachmen","Chevrolet","Hyundai",
                     "Kenworth","Case","Lexus","Nissan","Lincoln"," Mini Cooper","Dodge","Jayco","Heartland","Coleman","Indian",
                     "Audi"]
    for manufacturer in manufacturers:
        if manufacturer.lower() in model.lower():
            return manufacturer
    return None

# Apply the function to fill missing manufacturer values
df['manufacturer'] = df['manufacturer'].fillna(df['model'].apply(extract_manufacturer))
```

```python
# Function to correct specific patterns
df['model'] = df['model'].fillna('').astype(str)
def correct_patterns(text):
    text = re.sub(r'Toyta', 'Toyota', text)
    text = re.sub(r'Hondaa', 'Honda', text)
    text = re.sub(r'merceds', 'Mercedes', text)
    text = re.sub(r'voltswagon','Volkswagen',text)
    text = re.sub(r'Volkswagon','Volkswagen',text)
    text = re.sub(r'Chevy', 'Chevrolet', text)
    text = re.sub(r'2013 FOR', 'Ford', text)
    text = re.sub(r'LINCOL', 'Lincoln', text)
    text = re.sub(r'2001 AUDT', 'Audi', text)
    return text

# Apply to a column
df['model'] = df['model'].apply(correct_patterns)
```

```python
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
# Calculate average price by manufacturer and region
avg_price_by_manufacturer_region = df.groupby(['manufacturer', 'region'])['price'].mean().reset_index()
# Create a stacked bar chart
fig = px.bar(
    avg_price_by_manufacturer_region,
    x='manufacturer',
    y='price',
    color='region',  # Different colors for each region
    barmode='stack',  # Stacked bars
    title='Average Price by Manufacturer and Region',
    labels={'price': 'Average Price', 'manufacturer': 'Manufacturer'},
    text='price'
)

# Update layout
fig.update_traces(texttemplate='%{text:.2f}', textposition='inside')  # Place text inside bars
fig.update_layout(
    xaxis_title='Manufacturer',
    yaxis_title='Average Price',
    title_x=0.5
)

# Show the chart
fig.show()
```

```python
# Count vehicles by manufacturer
manufacturer_counts = df['manufacturer'].value_counts()

# Plot
plt.figure(figsize=(20, 6))
manufacturer_counts.plot(kind='bar', color='skyblue')
plt.title('Vehicle Count by Manufacturer')
plt.xlabel('Manufacturer')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
# Count vehicles by region
region_counts = df['region'].value_counts()

# Plot
plt.figure(figsize=(38, 8))
region_counts.plot(kind='bar', color='green')
plt.title('Vehicle Count by Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Remove potential outliers from the odometer column using the 1st and 99th percentiles
q1, q99 = df['odometer'].quantile(0.01), df['odometer'].quantile(0.99)
filtered_data = df[(df['odometer'] >= q1) & (df['odometer'] <= q99)]

# Plot the updated odometer distribution
plt.figure(figsize=(10, 6))
plt.hist(filtered_data['odometer'], bins=30, color='purple', edgecolor='black')
plt.title('Odometer Distribution (Outliers Removed)', fontsize=14)
plt.xlabel('Odometer', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Remove potential outliers from the price column using the 1st and 99th percentiles
q1_price, q99_price = df['price'].quantile(0.01), df['price'].quantile(0.99)
filtered_price_data = df[(df['price'] >= q1_price) & (df['price'] <= q99_price)]

# Plot the updated price distribution
plt.figure(figsize=(10, 6))
plt.hist(filtered_price_data['price'], bins=30, color='blue', edgecolor='black')
plt.title('Price Distribution (Outliers Removed)', fontsize=14)
plt.xlabel('Price', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
